```python
def plotModelResults1(prediction, x=None):
    """
        Plots modelled vs fact values, prediction intervals and anomalies
    """
    x = x
    prediction = prediction
    plt.figure(figsize=(15, 7))
    plt.plot(prediction, label="pediction", linewidth=2.0)

    plt.title(f'{x}')
```

## Пишем цикл для всех моделей

```python
preds_dict = {}

for csb in data['CSB'].unique():
    
    test_set = grouped_data[grouped_data['CSB'] == csb]
    control_test = test_set.drop(['CSB', 'ds', 'businessId'], axis=1)
    
    # Выводим название моделей xgb и etr
    filename_xgb_model = f'xgb_model_{csb}_2021-07-15.pkl'
    filename_etr_model = f'etr_model_{csb}_2021-07-15.pkl'
    
    #Загружаем модели
    clf_xgb = pickle.load(open(filename_xgb_model, 'rb'))
    clf_etr = pickle.load(open(filename_etr_model, 'rb'))
    
    # Делаем предсказания на год
    predict_xgb = clf_xgb.predict(control_test)
    predict_etr = clf_etr.predict(control_test)
    
    # Делаем предсказание по ансамблю (среднее по 2м моделям)
    predict_ens = np.mean([predict_xgb, predict_etr], axis=0)
    
    # Записываем предсказания в словарь
    preds_dict[csb] = predict_ens
    
    
```

```python
# Преобразуем данные в датафрейм
final_df = pd.DataFrame(preds_dict)
final_df['ds'] = yy
```
